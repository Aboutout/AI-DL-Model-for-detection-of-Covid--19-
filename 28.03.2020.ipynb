```python
import pandas as pd 
url="https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/03-28-2020.csv"
data = pd.read_csv(url,error_bad_lines=False) 
data.head()
```

```python
d=data.groupby(['Country_Region'],as_index=False).sum()
d1=d.sort_values(by='Confirmed',ascending=False)
```

```python
d1=d1.head(10)
d1
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

objects = d1['Country_Region']
y_pos = np.arange(len(objects))
performance = d1['Confirmed']

plt.barh(y_pos, performance, align='center', alpha=0.5)
plt.yticks(y_pos, objects)
plt.xlabel('Usage')
plt.title('Programming language usage')

plt.show()
```
